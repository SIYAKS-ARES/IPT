```python
#11.1
def sumColumn(matris, sutun_indeksi):
    toplam = 0
    for satir in matris:
        toplam += satir[sutun_indeksi]
    return toplam

def main():
    # Matrisin boyutlarını kullanıcıdan al
    satir_sayisi = int(input("Matrisin satır sayısını girin: "))
    sutun_sayisi = int(input("Matrisin sütun sayısını girin: "))

    # Boş bir matris oluştur
    matris = []

    # Matris elemanlarını kullanıcıdan al
    print("Matris elemanlarını satır satır girin:")
    for i in range(satir_sayisi):
        satir = []
        for j in range(sutun_sayisi):
            eleman = float(input(f"Eleman [{i}][{j}]: "))
            satir.append(eleman)
        matris.append(satir)

    # Matrisi ekrana yazdır
    print("\nMatris:")
    for satir in matris:
        print(satir)

    # Her sütunun toplamını hesapla ve ekrana yazdır
    for j in range(sutun_sayisi):
        sutun_toplami = sumColumn(matris, j)
        print(f"{j}. sütunun toplamı: {sutun_toplami}")

# Programı çalıştır
main()
```

```python
#11.6


def matrisCarp(a, b):
    # A matrisinin boyutlarını al
    m, n = len(a), len(a[0])
    # B matrisinin boyutlarını al
    p, q = len(b), len(b[0])

    # A matrisinin sütun sayısı ile B matrisinin satır sayısı aynı olmalı
    if n != p:
        print("Matrisler çarpılamaz. A matrisinin sütun sayısı B matrisinin satır sayısına eşit olmalı.")
        return None

    # Çarpım matrisini tutacak olan C matrisini oluştur
    c = [[0 for _ in range(q)] for _ in range(m)]

    # Çarpım matrisini hesapla
    for i in range(m):
        for j in range(q):
            for k in range(n):
                c[i][j] += a[i][k] * b[k][j]

    return c

# Kullanıcıdan matrisleri al
def matrisAl():
    matris = []
    satirSayisi = int(input("Matris satır sayısını girin: "))
    sutunSayisi = int(input("Matris sütun sayısını girin: "))
    print("Matris elemanlarını satır bazında girin:")

    for i in range(satirSayisi):
        satir = list(map(float, input().split()))
        if len(satir) != sutunSayisi:
            print("Girilen satırın eleman sayısı sütun sayısı ile uyuşmuyor.")
            return None
        matris.append(satir)

    return matris

# Kullanıcıdan iki matrisi al
print("İlk matrisi girin:")
matris1 = matrisAl()
print("İkinci matrisi girin:")
matris2 = matrisAl()

# Matrisleri çarp ve sonucu ekrana yazdır
if matris1 and matris2:
    sonuc = matrisCarp(matris1, matris2)
    if sonuc:
        print("Çarpım matrisi:")
        for satir in sonuc:
            print(satir)
```

```python
#11.9


def oyun_tahtasi_goster(tahta):
    # Tic-tac-toe oyun tahtasını gösterir
    for row in tahta:
        print("|".join(row))
        print("-" * 5)

def kazanan_mi(tahta, oyuncu):
    # Tahtada kazananın olup olmadığını kontrol eder
    for i in range(3):
        if all(tahta[i][j] == oyuncu for j in range(3)) or \
           all(tahta[j][i] == oyuncu for j in range(3)):
            return True
    if all(tahta[i][i] == oyuncu for i in range(3)) or \
       all(tahta[i][2 - i] == oyuncu for i in range(3)):
        return True
    return False

def bos_hucre_var_mi(tahta):
    # Tahtada boş hücre olup olmadığını kontrol eder
    return any("_" in row for row in tahta)

def hamle_yap(tahta, oyuncu, satir, sutun):
    # Oyuncunun hamlesini tahtaya uygular
    if tahta[satir][sutun] == "_":
        tahta[satir][sutun] = oyuncu
        return True
    return False

def main():
    # Oyunun ana fonksiyonu
    tahta = [["_"] * 3 for _ in range(3)]
    oyuncular = ["X", "O"]
    siradaki_oyuncu = 0
    devam = True

    while devam:
        oyun_tahtasi_goster(tahta)
        oyuncu = oyuncular[siradaki_oyuncu]
        print(f"{oyuncu} oyuncusu sırası. (Satır ve sütun numarasını aralarında boşluk bırakarak girin)")

        try:
            satir, sutun = map(int, input().split())
            if satir < 0 or satir > 2 or sutun < 0 or sutun > 2:
                raise ValueError
        except ValueError:
            print("Geçersiz hamle. Lütfen satır ve sütun numaralarını 0, 1 veya 2 arasında girin.")
            continue

        if not hamle_yap(tahta, oyuncu, satir, sutun):
            print("Bu hücre dolu. Lütfen başka bir hücre seçin.")
            continue

        if kazanan_mi(tahta, oyuncu):
            oyun_tahtasi_goster(tahta)
            print(f"{oyuncu} oyuncusu kazandı!")
            devam = False
        elif not bos_hucre_var_mi(tahta):
            oyun_tahtasi_goster(tahta)
            print("Berabere!")
            devam = False

        siradaki_oyuncu = (siradaki_oyuncu + 1) % 2

main()
```

```python
#11.11
def decimal_to_binary(decimal):
    binary = bin(decimal)[2:].zfill(9)  # 9-bit binary representation
    return binary

def print_matrix(binary):
    for i in range(0, 9, 3):
        row = binary[i:i+3]
        row = row.replace('0', 'H')  # Heads
        row = row.replace('1', 'T')  # Tails
        print(row)

def main():
    decimal = int(input("0 ile 511 arasında bir sayı girin: "))
    if not 0 <= decimal <= 511:
        print("Geçersiz giriş! Lütfen 0 ile 511 arasında bir sayı girin.")
        return

    binary = decimal_to_binary(decimal)
    print("Karakter Matrisi:")
    print_matrix(binary)

# Programı çalıştır
main()
```

```python
#11.13
def locateLargest(a):
    max_value = a[0][0]
    max_row, max_col = 0, 0

    for i in range(len(a)):
        for j in range(len(a[i])):
            if a[i][j] > max_value:
                max_value = a[i][j]
                max_row, max_col = i, j

    return [max_row, max_col]

def main():
    rows = int(input("Matrisin satır sayısını girin: "))
    cols = int(input("Matrisin sütun sayısını girin: "))

    matrix = []
    for i in range(rows):
        row = input(f"{i+1}. satırı girin (virgülle ayırın): ").split(',')
        row = [int(num) for num in row]
        matrix.append(row)

    largest_location = locateLargest(matrix)
    print("En büyük elemanın konumu:", largest_location)

# Programı çalıştır
main()
```

```python
#11.14
import random

def print_matrix(matrix):
    for row in matrix:
        print(' '.join(map(str, row)))

def check_all_same(lst):
    return all(elem == lst[0] for elem in lst)

def find_all_same_rows(matrix):
    all_same_rows = []
    for i, row in enumerate(matrix):
        if check_all_same(row):
            all_same_rows.append(i)
    return all_same_rows

def find_all_same_cols(matrix):
    all_same_cols = []
    for j in range(len(matrix[0])):
        column = [matrix[i][j] for i in range(len(matrix))]
        if check_all_same(column):
            all_same_cols.append(j)
    return all_same_cols

def find_main_diagonal(matrix):
    main_diagonal = [matrix[i][i] for i in range(len(matrix))]
    if check_all_same(main_diagonal):
        return True
    return False

def main():
    size = int(input("Kare matrisin boyutunu girin: "))
    matrix = [[random.randint(0, 1) for _ in range(size)] for _ in range(size)]

    print("Rastgele oluşturulan kare matris:")
    print_matrix(matrix)

    same_rows = find_all_same_rows(matrix)
    if same_rows:
        print("Tüm 0'lardan veya tüm 1'lerden oluşan satırların indeksleri:", same_rows)
    else:
        print("Herhangi bir tüm 0'lardan veya tüm 1'lerden oluşan satır bulunamadı.")

    same_cols = find_all_same_cols(matrix)
    if same_cols:
        print("Tüm 0'lardan veya tüm 1'lerden oluşan sütunların indeksleri:", same_cols)
    else:
        print("Herhangi bir tüm 0'lardan veya tüm 1'lerden oluşan sütun bulunamadı.")

    if find_main_diagonal(matrix):
        print("Ana diyagonalde tüm 0'lar veya tüm 1'ler var.")
    else:
        print("Ana diyagonalde tüm 0'lar veya tüm 1'ler yok.")

# Programı çalıştır
main()
```

```python
#11.15
def sameLine(points):
    if len(points) < 3:
        return True
    x0, y0 = points[0]
    x1, y1 = points[1]
    for i in range(2, len(points)):
        x, y = points[i]
        if (x1 - x0) * (y - y0) != (x - x0) * (y1 - y0):
            return False
    return True

def main():
    points = []
    for i in range(5):
        x, y = map(float, input(f"Lütfen {i+1}. noktanın x ve y koordinatlarını girin (örn. 1 2): ").split())
        points.append((x, y))

    if sameLine(points):
        print("Girilen noktaların hepsi aynı doğru üzerinde.")
    else:
        print("Girilen noktalar aynı doğru üzerinde değil.")

# Programı çalıştır
main()
```
